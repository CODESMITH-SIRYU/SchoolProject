오버로딩은 + 라는 연산자에 대해서 int + int 로만 구현이 되어 있었다면, 1 + 1.0f 와 같이 int + float역시 대응할 수 있게 만드는 것이다.  
다른 말로는 중복 정의라고 하는데, 파이썬은 중복 정의 할 경우 가장 마지막에 정의한 코드로 실행된다.
오버라이딩은 연산자나 함수를 재정의하여 처리와 결과값을 바꾸는 것이다. 예를 들면 원과 정사각형의 둘레를 구하는 공식은 다르지만,  
둘레를 구한다는 점은 똑같다. 이런 때에 오버라이딩을 사용하여 처리를 달리 할 수 있다. 오버라이딩은 재정의라고 한다.

```python
# 오버로딩에 관한 코드

def m_add(a, b):
    return a+b

def m_add(a,b):
    return a+b+1

# 오버라이딩에 관한 코드

class polygons:
    def area(self):
        pass

class angled_triangle(polygons):
    def __init__(self,a, b, c):  # 세 변의 길이에 제한이 있지만 여기서는 고려하지 않는다.
                            # a,b는 밑변과 높이이고, c는 빗변이다.
        self.a = a
        self.b = b
        self.c = c
    def area(self):
        return self.a*self.b/2
    
class rectangle(polygons):
    def __init__(self,a,b): # 직사각형의 마주보는 두 변은 같으므로 두 변의 길이만 알면 된다.
        self.a = a
        self.b = b
    def area(self):
        return self.a*self.b
```

### [문제 3]

다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)  

* 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.  
  
    count: Counter 인스턴스가 지니는 초기 정수 값  
    step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)  
    \>>> c = Counter(10)   
    \>>> d = Counter(10, 2)   
    
      
* 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.   

    \>>> print(c)   
    [Count (step: 1)] 10   
    \>>> print(d)   
    [Count (step: 2)] 10   
    
    
* 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오. 

    \>>> c.incr()   
    \>>> d.incr()   
    \>>> print(c)   
    [Count (step: 1)] 11   
    \>>> print(d)   
    [Count (step: 2)] 12   
    
      
* 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.  

    \>>> c()   
    \>>> d()   
    \>>> print(c)   
    [Count (step: 1)] 12   
    \>>> print(d)   
    [Count (step: 2)] 14   
    
      
* 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.   

    \>>> c = c + 5   
    \>>> d = d - 5   
    \>>> print(c)   
    [Count (step: 1)] 17   
    \>>> print(d)   
    [Count (step: 2)] 9   
    
    
* 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.  

    \>>> print(c > 10)  
    True   
    \>>> print(d > 10)   
    False   
    \>>> print(c < 10)   
    False   
    \>>> print(d < 10)   
    True   
    \>>> print(c == 17)   
    True    
    \>>> print(d != 9)   
    False   

```python
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
    def __str__(self):
        return "[Count (step: %s)] %s" % (self.step, self.count)
    def incr(self):
        self.count = self.count + self.step
    def __call__(self):
        self.incr()
    def __add__(self, other):
        self.count = self.count + other
        return self
    def __sub__(self, other):
        self.count = self.count - other
        return self
    def __lt__(self, other):
        return self.count < other
    def __gt__(self, other):
        return self.count > other
    def __eq__(self, other):
        return self.count == other
```

```python
c = Counter(10, 4)
print(c)
c()
print(c)
c = c + 5
print(c)
c = c - 5
print(c)
print(c < 10)
print(c > 10)
print(c == 10)
```

### [문제 4]

다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \_\_init__(), \_\_str__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

Myset은 class 생성자에서 부터 list를 상속한다고 정의가 되어 있기 때문에, 우선 list와 동일한 자격, 함수, 연산자를 가진다.  
여기서 \_\_init__ 으로 생성자에서 l을 입력받는데, l의 요소들을 Myset에 append시킨다.  
이는 Myset이 list를 상속받았기 때문에 사용가능하다.  
append시킨 이후에는 Myset의 정적 메소드인 eliminate_duplicate인데, 이는 MySet의 클래스 함수가 아니다.
인자도 self가 없기 때문에 Myset의 이름 공간 안에 존재하기만 하는 함수이면서, 정적 메소드이기 때문에 인스턴트를 생성하지 않아도  
호출이 가능하다.  
eliminate_duplicate는 내부에서 빈 리스트 s를 생성하고, 인자로 받은 l의 요소들을 조회하면서 s에 추가하는데, 이 때, s에 중복된  
내용이 있는 지 없는 지 검사하여 중복된 내용은 배제하고 append한다.  
그리고 인자로 받은 l을 비우고, s의 요소들 다시 l에 복사하여, 결론적으로 l 내에 있는 요소의 중복을 제거한다.  
\_\_str__은 Myset이 string으로 캐스팅 될 때에 호출되는 함수인데,  
result에 "Myset: {"에 요소들을 요소 + " ," 형식으로 붙이고, result = result\[0:len(result)-2\] + "}"로,  
가장 마지막 요소 뒤에 붙는 " ,"를 문자열 슬라이싱으로 잘라내고 중괄호를 닫아,  
최종적으로는 "Myset: {요소 ,요소 , 요소}"와 같은 형식으로 반환을 한다.

### [문제 5]

4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오  

정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.  

- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.  

    \>>> u = s | t   
    \>>> print(u)   
    MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}   
      
      
- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.  

    \>>> u = s & t   
    \>>> print(u)   
    MySet: {2, 3}   
        
        
- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.  
  
    \>>> s = MySet([1, 2, 3])   
    \>>> t = MySet([3, 4, 5])   
    \>>> u = s - t   
    \>>> print(u)   
    MySet: {1, 2}   
    

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __and__(one, other): # 교집합을 반환하는 연산자 재정의
        l = []
        for e in one:
            if e in other:
                l.append(e)  # A에도 있고, B에도 있는 것이 교집합이므로
                             # if e in other를 만족하는 것 만을 l에 append하고
                             # l를 MySet으로 생성하여 반환한다
        return MySet(l)
    def __or__(one, other): # 합집합을 반환하는 연산자 재정의
                            # 두 집합을 합치고 중복을 제거한 것이 합집합이기 때문에
                            # 정적 메소드로 정의된 eliminate_duplicate 재활용
                            # l을 MySet으로 생성하는 과정에서 eliminate_duplicate를 사용하므로
                            # l은 그저 one + other를 하고 Myset(l)로 생성하여 반환한다.
        l = one + other
        return MySet(l)
    def __sub__(one, other): # 차집합을 반환하는 연산자 재정의
                             # one에 있는 것 중에서 other에는 없는 것만을(if e not in other) 반환한다.
                             # 이후 l을 MySet으로 생성하여 반환한다.
        l = []
        for e in one:
            if e not in other:
                l.append(e)
        return MySet(l)
```

```python
s = MySet([3,4,5,6,7,8])
t = MySet([7,8,9,10,11])
c = s | t
print(c)
c = s & t
print(c)
c = s - t
print(c)
```

```python
class SearchEngine:
    
    def __init__(self, *link):
        
        # 불용어 로딩
        self.stop = self.loadStopWords()
        # 각 링크의 단어 출현 빈도수를 저장할 리스트, 튜플(링크, 빈도수를 가진 사전)을 가지는 리스트
        self.lists = []
        
        for e in link:
            # 링크와 링크의 빈도수를 리스트에 저장한다
            self.lists.append((e, self.getFrequencyFromUrl(e)))
            
    def addUrl(self, link):
        # 리스트의 튜플에 링크와 빈도수를 저장한다
        self.lists.append((link, self.getFrequencyFromUrl(link)))

    def removeUrl(self, link):
        # 리스트의 튜플에서 link와 내용을 삭제한다
        for (e, k) in self.lists:
            if link in e:
                self.lists.remove((e, k))
        
    def listUrls(self):
        # 리스트의 튜플에서 앞 값들만을 불러온다
        for (e, k) in self.lists:
            print(e)
    
    def getWordsFrequency(self):
        # 리스트의 튜플들을 전부 출력하면 끝이다
        for (e, k) in self.lists:
            print(k)
        
    def getMaxFrequencyWord(self):
        # 최대 빈도수를 저장할 변수
        big_v = -1
        # url들을 저장할 변수
        big_k = []
        for (i,e) in self.lists: # (link, 빈도수를 나타내는 사전)을 가리키는 반복자 i,e
            for j in e.keys(): # 어떤 링크의 사전의 key를 나타내는 반복자
                        # 그 링크의 value값을 전부 검사하면서 현재 최대 빈도수(big_v) 보다 크면 그 단어와 값을
                        # big_k와 big_v에 저장
                        # 같으면 big_k에 추가만 한다
                if (e[j] > big_v):
                    big_k = [j]
                    big_v = e[j]
                elif (e[j]==big_v):
                    big_k.append(j)
        print(big_k)
        
    def searchUrlByWord(self, word):
        word = word.lower()
        # url들을 저장하는 변수
        urls = []
        # 최대 유사도를 저장하는 변수
        simil = 0
        for (i,e) in self.lists: # (link, 빈도수를 나타내는 사전)을 가리키는 반복자 i,e
            for j in e.keys(): # 어떤 링크의 사전의 key를 나타내는 반복자 j
                        # word가 j안에 있으면 그 유사도가 현재 최대 유사도보다 큰 지 검사하고,
                        # 최대 유사도보다 크면 urls에 그 url(i.key())을 저장하고 현재 최대 유사도에 유사도를 대입
                        # 만약 유사도가 같으면 urls에 현재 url을 추가
                if word in j:
                    if (e[j] > simil):
                        urls = [i]
                        simil = e[j]
                    elif (e[j] == simil):
                        urls.append(i)
        print(urls)
                    
    @staticmethod
    def loadStopWords():
        
        # 특수문자들을 제거하기 위한 불용어 로딩
        stop_en = []
        stop_kr = []
        with open("./files/stopword_en.txt", "r") as f:
            line = f.readline()
            while(line):
                stop_en.append(line.strip("\n"))
                line = f.readline()
        with open("./files/stopword_kr.txt", "r") as f:
            line = f.readline()
            while(line):
                stop_kr.append(line.strip("\n"))
                line = f.readline() 
    
        stop = stop_en + stop_kr
        return stop
    
    def getFrequencyFromUrl(self, link):
        
        import requests
        import re
        import string
        
        # 소스를 가져온다
        req = requests.get(link)
        #소스를 텍스트로 변환한다
        source = req.text
        # 정규 표현식으로 단어들 가져오기
        text1 = re.sub('<script.*?>.*?</script>', ' ', source, 0, re.I|re.S) 
        text2 = re.sub('<style.*?>.*?</style>', ' ', text1, 0, re.I|re.S)    
        text3 = re.sub('<.+?>', ' ', text2, 0, re.I|re.S)                    
        text4 = re.sub('-->', ' ', text3, 0, re.I|re.S)     
        text5 = re.sub("[!|?|<|>|:|\[|\]|#|$|%|&|\(|\)|*|+|\-|,|.|/|;|=|@|^|_|`|{|}|~|\"|\'|\\\|/|★]",' ', text4, 0, re.I|re.S)

        s = text5.split()
        
        # 불용어를 제외한 단어들을 사전에 저장하게 한다.
        dic = dict()

        for i in s:
            i = i.lower() # 모든 불용어는 소문자이기 때문에 영어일 경우 소문자로 저장해야 한다.
            if(i in dic):
                dic[i] = dic[i] + 1
            elif((i not in dic)and(i not in self.stop)):
                dic[i] = 1
        
        return dic
```

```python
w1 = SearchEngine('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com')
```

```python
w1.listUrls()
print()
w1.addUrl('https://github.com')
w1.removeUrl('http://www.times.com')
w1.listUrls()
print()
w1.getWordsFrequency()
print()
w1.getMaxFrequencyWord()
print()
w1.searchUrlByWord("news")
```

### [8번 문제]

위 7번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된   
getWordsFrequency() 메소드를 오버라이드하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.  
  
    리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.  
    getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.  
  
    reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.  
    reverse 인자에 true를 넣으면 오름차순으로 정렬한다.  
        >>> w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')  
  
        >>> w4.getWordsFrequency()  
        [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]  
  
        >>> w4.getWordsFrequency(reverse=true)  
        [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]  

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    def getWordsFrequency(self, reverse=False):
        self.l = []
        l2 = []
        for (i,e) in self.lists: # list의 (사이트, 사이트의 단어 출현빈도 사전)
            for j in e.keys():   # 단어 출현 빈도 사전의 key값들
                self.l.append((e[j], j)) # 리스트에 숫자-단어 순으로 밀어넣는다
        # 리스트를 정렬한다. 단, 기본이 오름차순이기 때문에 not reverse로 정렬시킨다
        self.l.sort(reverse = not reverse)
        # 리스트를 정렬한 다음에는 다시 단어-숫자 순으로 다른 리스트에 밀어넣는다
        for (e,j) in self.l:
            l2.append((j, e))
        # self.l에 순서가 바뀐 리스트를 가리키게 한다
        self.l = l2
        print(self.l)
    def __len__(self):
        return len(self.l)
    
    def __getitem__(self, k):
        if not ((k < len(self)) or (k>0)):
            raise (IndexError, k)
        return self.l[k]
```

```python
w2 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
```

```python
w2.getWordsFrequency()
w2.getWordsFrequency(reverse=True)
```
